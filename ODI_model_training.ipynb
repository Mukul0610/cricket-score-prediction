```python
import pandas as pd
import numpy as np
import pickle
```

```python
df=pickle.load(open('final_data.pkl','rb'))
```

```python
df['score']=df.groupby('match_id').cumsum()['runs']
```

```python
df['out']=np.where(df['out'].str.isnumeric(),0,1)
```

```python
df['num_of_wickets']=df.groupby('match_id').cumsum()['out']
```

```python
df['wicket_left']=10-df['num_of_wickets']
```

```python
df=df.drop(columns=['out','num_of_wickets'])
```

```python
df['over']=df['ball'].astype(int)
```

```python
df['ball_num']=(df['ball']-df['over'])*10
```

```python
df['ball_num']=df['ball_num'].astype(int)
```

```python
df['ball_bowled']=(df['over']*6+df['ball_num']).astype(int)
```

```python
df['ball_left']=300-df['ball_bowled']
```

```python
df['crr'] = (df['score']*6)/df['ball_bowled']
```

```python
df
```

```python
match_ids=df['match_id'].unique()
last_ten=[]
for i in match_ids:
    last_ten.extend(df.groupby('match_id').get_group(i).rolling(window=60).sum()['runs'].values.tolist())
```

```python
df['last_10_over']=last_ten
```

```python
df=df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')
```

```python
df.dropna(inplace=True)
```

```python
df=df[['batting_team','bowling_team','city','score','ball_left','wicket_left','crr','last_10_over','runs_x']]
```

```python
df=df.sample(df.shape[0])
```

```python
X = df.drop(columns=['runs_x'])
y = df['runs_x']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
tr1=ColumnTransformer([
    ('Ohe',OneHotEncoder(),['batting_team','bowling_team','city'])
],remainder='passthrough')
```

```python
pipe=Pipeline([
    ('pipe1',tr1),
    ('pipe2',StandardScaler(with_mean=False)),
    ('pipe3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
pipe.fit(X_train,y_train)
```

```python
x=pipe.predict(X_test)
```

```python
r2_score(y_test,x)
```

```python
pickle.dump(pipe,open('pipe_odi.pkl','wb'))
```
