```python
import pickle
import pandas as pd
import numpy as np
```

```python
df=pickle.load(open('df.pkl','rb'))
```

```python
df=df[df['info.gender']=='male']
```

```python
df=df.drop(columns=['info.gender',
                 'info.match_type',
                 'info.outcome.by.runs',
                 'info.outcome.winner',
                 'info.supersubs.Asia XI',
                 'info.supersubs.Zimbabwe',
                 'info.supersubs.Sri Lanka',
                 'info.supersubs.Bangladesh',
                 'info.supersubs.New Zealand',
                 'info.supersubs.India',
                 'info.supersubs.ICC World XI',
                 'info.supersubs.South Africa',
                 'info.supersubs.Pakistan',
                 'info.supersubs.West Indies',
                 'info.outcome.by.wickets',
                 'info.outcome.result',
                 'info.outcome.method',
                 'info.neutral_venue',
                 'info.match_type_number',
                 'info.outcome.eliminator',
                 'info.supersubs.Australia',
                 'info.supersubs.England',
                 'info.supersubs.Africa XI',
                'info.dates',
                 'info.overs',
                 'info.player_of_match',
                'meta.revision',
                'meta.data_version',
                 'meta.created',
                'info.toss.decision',
                 'info.toss.winner',
                'info.umpires'
                   ])
```

```python
df
```

```python
count=0
final_df=pd.DataFrame()
for index,row in df.iterrows():
    if count in [8,60,64,87,114,121,224,265,309,359,396,412,500,565,647,674,690,721,736,739,759,760,772,828,904,906,917,980,1032,
                1052,1152,1160,1162,1266,1271,1308,1326,1372,1395,1432,1621,1627,1630,1653,1664,1740,1768]:
        count+=1
        continue
    count+=1
    match_id=[]
    ball_no=[]
    batting_teams=[]
    bowling_teams=[]
    run=[]
    player_out=[]
    city=[]
    venue=[]
    for ball in row['innings'][0]['1st innings']['deliveries']:
        for key in ball.keys():
            ball_no.append(key)
            match_id.append(count)
            batting_teams.append(row['innings'][0]['1st innings']['team'])
            bowling_teams.append(row['innings'][1]['2nd innings']['team'])
            city.append(row['info.city'])
            venue.append(row['info.venue'])
            run.append(ball[key]['runs']['total'])
            try:
                player_out.append(ball[key]['wicket']['player_out'])
            except:
                player_out.append('0')
    loop_df=pd.DataFrame({
        'match_id':match_id,
        'batting_team':batting_teams,
        'bowling_team':bowling_teams,
        'runs':run,
        'ball':ball_no,
        'out':player_out,
        'city':city,
        'venue':venue
    })
    final_df=final_df.append(loop_df)
```

```python
df=final_df.copy()
```

```python
team=['Australia','Sri Lanka','India','England','Pakistan','South Africa','New Zealand','West Indies','Bangladesh',
      'Zimbawe','Afghanistan','Ireland']
```

```python
df=df[df['batting_team'].isin(team)]
```
